```python
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
salary = pd.read_csv("/content/Salary.csv")
salary
```

```python
x = salary["YearsExperience"]
y = salary["Salary"]
```

```python
plt.scatter(x,y)
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
```

```python
from sklearn.model_selection import train_test_split #Fixed typo 'tarin_test_split' to 'train_test_split'
x_train, x_test , y_train , y_test = train_test_split(x, y , test_size = 0.2 , random_state = 0) # Changed 'random' to 'random_state'
```

```python
x_train = np.array(x_train).reshape(-1,1) # Reshape to have 28 rows and 1 column
x_test = np.array(x_test).reshape(-1,1) # Reshape to have number of rows in x_test and 1 column
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_train)
```

```python
plt.scatter(x_train,y_train, color = "blue")
plt.plot(x_train, y_pred, color = "red")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x_test,y_test, color = "blue")
plt.plot(x_train, y_pred, color = "red")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```
